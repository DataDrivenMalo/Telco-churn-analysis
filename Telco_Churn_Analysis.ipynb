```python
#Feature Engineering
#handle missing values. Fill with median or mean
data['TotalCharges'].fillna(data['TotalCharges'].median(),inplace=True)
data
```

```python
#remove duplicates
data.drop_duplicates()
```

```python
#Label & one hot encoding
from sklearn.preprocessing import LabelEncoder

#label encode binary columns
#remove duplicates
binary_cols = ['Churn','Partner','Dependents','PhoneService','OnlineSecurity','DeviceProtection','TechSupport']
le = LabelEncoder()
for col in binary_cols:
    data[col] = le.fit_transform(data[col])

#one hot encode multiclass categorical features
data = pd.get_dummies(data, columns = ['InternetService','Contract','PaymentMethod'],drop_first=True)
```

```python
data.head()
```

```python
#Calculate Mutual Info scores to see which features are informative
from sklearn.feature_selection import mutual_info_classif

X = data.drop(columns = 'Churn').select_dtypes(exclude='object')
y = data['Churn']
X= X.dropna()
y = y.loc[X.index]
mi_scores = mutual_info_classif(X,y,discrete_features = 'auto')
mi_scores = pd.Series(mi_scores,index=X.columns).sort_values(ascending=False)
print(mi_scores)
```

```python
#Drop low MI features
low_mi = mi_scores[mi_scores<0.02].index.tolist()
X.drop(columns=low_mi,inplace=True)
```

```python
#Feature creation
data['OnlineBackup'] = data['OnlineBackup'].map({'Yes' : 1,'No':0})
data['StreamingTV'] = data['StreamingTV'].map({'Yes': 1,'No':0})
data['StreamingMovies'] = data['StreamingMovies'].map({'Yes' : 1,'No' : 0})
data['InternetService_Fiber optic'] = data['InternetService_Fiber optic'].map({'False':0,'True':1})
data['InternetService_No'] = data['InternetService_No'].map({'False':0,'True':1})
data
```

```python
X['TotalServices'] = (data[['PhoneService','OnlineSecurity','InternetService_Fiber optic','InternetService_No','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']]).sum(axis=1)
X.head()
```

```python
X['PricePerService'] = X['MonthlyCharges']/(X['TotalServices'].replace(0,1)) #use replace to avoid division by 0
X['TenureBucket'] = pd.cut(data['tenure'],bins=[0,12,24,48,60,100],labels=['0-12','12-24','24-48','48-60','60+'])
X['IsAutoPay'] = data['PaymentMethod_Credit card (automatic)'].astype(int)
X.head()
```

```python
#to handle skewness,do log transformation
X['LogMonthlyCharges'] = np.log1p(X['MonthlyCharges'])
X['LogTotalCharges'] = np.log1p(X['TotalCharges'])
X
```

```python
#Interaction Charges
X['Services_Charges']=X['TotalServices']*X['MonthlyCharges']
```

```python
#Scaling(normalization) for numeric features
numeric_features = ["tenure",'MonthlyCharges','TotalCharges','PricePerService','LogMonthlyCharges','LogTotalCharges','Services_Charges']
#list comprehension to check if the above numeric features are present in X
numeric_features = [f for f in numeric_features if f in X.columns]
```

```python
#Imbalance Handling
#in classification models, target classes might be not evenly distributed(class imbalance). To balance the dataset,we need to either undersample (reduce the number of majority class samples) or oversample(increase the number of minority class sample)
#SMOTE- Synthetic minority over-sampling technique. It won't duplicate the majority class as that might cause overfitting.SMOTE generates synthetic data points by interpolating real ones. it creates new, slightly different versions of existing minority samples. It only takes numerical features

from imblearn.over_sampling import SMOTE
import pandas as pd

#drop tenor bucket from features
X_no_tenure_bucket = X.drop(columns=['Tenure_bucket'],errors='ignore')
#one hot encode the categorical variables
X_encoded = pd.get_dummies(X_no_tenure_bucket,drop_first=True)

#Apply SMOTE
sm = SMOTE(random_state=42)
X_resampled, y_resampled = sm.fit_resample(X_encoded,y)
```

```python
#Split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_resampled,y_resampled,test_size=0.2,random_state=42,stratify=y_resampled)
```

```python
#train two models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

#Logistic Regression
#gradient descent - optimization algorithm used to minimize loss function
#loss function - way to measure how wrong your model's predictions are.Lower the loss, better the model is doing
lr_model = LogisticRegression(max_iter=1000,random_state=42)
lr_model.fit(X_train,y_train)

#Random forest
#estimatores mean the number of trees. more trees mean less variance.but more trees = more training and memory time
rf_model = RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(X_train,y_train)

```

```python
#evaluate both the models
#we'll look at Accuracy, precision, recall,F1 Score,ROC AUC, Confusion matrix
#Precision - out of all predicted positives,how many were actually correct?
#Recall - out of all actual positives, how many did the model catch?

from sklearn.metrics import classification_report,confusion_matrix,roc_auc_score

#this step is to assess how confident the model is that each sample is class 1
def evaluate_model(model, X_test, y_test, name="Model"):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]

    print(f"--- {name} ---")
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
    print("\n")

#Evaluate logistic regression
evaluate_model(lr_model,X_test,y_test,name="Logistic_Regression")

evaluate_model(lr_model,X_test,y_test,name="Random_Forest")


```

```python
#Check for both models
#model 1
lr_model.fit(X_train,y_train)
evaluate_model(lr_model,X_test,y_test,name="Logistic Regression")

#model 2
rf_model.fit(X_train, y_train)
evaluate_model(rf_model,X_test,y_test,name='Random Forest')
```

```python
#Since both models are generalizing to the test set very well,try k-cross validation.It divides the dataset into k folds, trained on k-1 folds and test on remaining fold. repeat this k times and then average to provide a robust estimate of model's performance


from  sklearn.model_selection import cross_val_score
def compare_models_cv(models,X,y,cv=5,scoring='roc_auc'):
    """
    Compare multiple models using cross-validation.

    Parameters:
    - models: dict of model name -> model instance
    - X: feature set
    - y: target
    - cv: number of folds
    - scoring: metric to evaluate ('accuracy', 'f1', 'roc_auc', etc.)
    """
    results ={}
    print(f"scoring method:{scoring}")
    print("_"*40)

    for name,model in models.items():
        scores = cross_val_score(model,X,y,cv=cv,scoring=scoring)
        results[name] = scores
        print("f{name")
        print(f' Scores: {np.round(scores,4)}')
        print(f"Mean:{scores.mean():4f},Std Dev:{scores.std():.4f}")
        print()
```

```python
models = {
    "LogisticRegression":LogisticRegression(max_iter = 1000,random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=100,random_state=42)
}

compare_models_cv(models,X_resampled,y_resampled,cv=5,scoring='roc_auc')
```
